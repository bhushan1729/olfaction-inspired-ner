# Baseline vs Olfactory NER Experiments

**Hypothesis**: Does my biologically inspired olfactory feature extractor add value beyond a standard language model for NER?

## Experimental Design

### Baseline (LLM-only)
```
mBERT (frozen) → Linear Classifier → NER Tags
```

### Olfactory (LLM + Bio-Inspired)
```
mBERT (frozen) → Receptors → Glomeruli → BiLSTM → CRF → NER Tags
```

2. WikiANN Hindi - Low resource
3. WikiANN Marathi - Low resource

5. WikiANN Bangla - Low resource

# Analysis

Now that all experiments are complete, let's analyze the results.

## Quick Results Summary

````python
import json
import pandas as pd
from pathlib import Path

# Collect all results
results_data = []

datasets_info = [
    ('conll2003', None, 'en', 'CoNLL-2003 (English)'),
    ('wikiann', 'hi', 'hi', 'WikiANN Hindi'),
    ('wikiann', 'mr', 'mr', 'WikiANN Marathi'),
    ('wikiann', 'ta', 'ta', 'WikiANN Tamil'),
    ('wikiann', 'bn', 'bn', 'WikiANN Bangla'),
    ('wikiann', 'te', 'te', 'WikiANN Telugu'),
]

for dataset_name, lang_code, result_lang, display_name in datasets_info:
    for exp_type in ['baseline', 'olfactory']:
        result_path = Path(SAVE_DIR) / dataset_name / result_lang / f"mbert_{exp_type}" / "results.json"
        
        if result_path.exists():
            with open(result_path) as f:
                metrics = json.load(f)
            
            results_data.append({
                'Dataset': display_name,
                'Model': exp_type.capitalize(),
                'F1': metrics.get('test_f1', 0.0),
                'Precision': metrics.get('test_precision', 0.0),
                'Recall': metrics.get('test_recall', 0.0),
            })

# Create DataFrame
df = pd.DataFrame(results_data)

# Pivot for comparison
comparison = df.pivot_table(
    index='Dataset',
    columns='Model',
    values=['F1', 'Precision', 'Recall']
)

# Calculate improvements
comparison[('Improvement', 'F1')] = comparison[('F1', 'Olfactory')] - comparison[('F1', 'Baseline')]
comparison[('Improvement', 'Precision')] = comparison[('Precision', 'Olfactory')] - comparison[('Precision', 'Baseline')]
comparison[('Improvement', 'Recall')] = comparison[('Recall', 'Olfactory')] - comparison[('Recall', 'Baseline')]

print("\n" + "="*100)
print("BASELINE VS OLFACTORY: RESULTS SUMMARY")
print("="*100)
print(comparison.to_string())
print("="*100)

# Save to Drive
comparison.to_csv(f"{DRIVE_PATH}/quick_results_summary.csv")
print(f"\n✓ Results saved to {DRIVE_PATH}/quick_results_summary.csv")
````

## Statistical Significance

````python
# Load and display statistical tests
with open(f"{DRIVE_PATH}/comparison_analysis/statistical_tests.json") as f:
    stats = json.load(f)

print("\n" + "="*80)
print("STATISTICAL SIGNIFICANCE TESTS")
print("="*80)

if 'paired_t_test' in stats:
    t_test = stats['paired_t_test']
    print(f"\nPaired t-test:")
    print(f"  t-statistic: {t_test['t_statistic']:.4f}")
    print(f"  p-value: {t_test['p_value']:.4f}")
    print(f"  Significant (α=0.05): {t_test['significant']}")
    print(f"  → {t_test['interpretation']}")

if 'effect_size' in stats:
    effect = stats['effect_size']
    print(f"\nEffect Size (Cohen's d):")
    print(f"  d = {effect['cohens_d']:.4f}")
    print(f"  Interpretation: {effect['interpretation']}")

print("="*80)
````

## Final Summary

````python
# Calculate overall statistics
baseline_f1s = comparison[('F1', 'Baseline')].values
olfactory_f1s = comparison[('F1', 'Olfactory')].values
improvements = comparison[('Improvement', 'F1')].values

wins = (improvements > 0).sum()
total = len(improvements)

print("\n" + "="*80)
print("FINAL SUMMARY")
print("="*80)
print(f"\nDatasets where Olfactory wins: {wins}/{total} ({wins/total*100:.1f}%)")
print(f"\nAverage Baseline F1:   {baseline_f1s.mean():.4f} (±{baseline_f1s.std():.4f})")
print(f"Average Olfactory F1:  {olfactory_f1s.mean():.4f} (±{olfactory_f1s.std():.4f})")
print(f"Average Improvement:   {improvements.mean():.4f} (±{improvements.std():.4f})")

if wins >= total * 0.67 and improvements.mean() > 0:
    print("\n✅ HYPOTHESIS VALIDATED")
    print("The olfactory feature extractor adds value beyond standard mBERT for NER!")
else:
    print("\n⚠️ RESULTS MIXED")
    print("Further investigation needed.")

print("\n" + "="*80)
print(f"All results saved to Google Drive:")
print(f"  {DRIVE_PATH}")
print("="*80)
````
